```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_excel('Birmingham.xlsx')
```

```python
df = df.pivot('Location', 'Category', 'Number of Tweets')
plt.figure(figsize=(15,10))
g=sns.heatmap(df,annot=True, linewidth = 0.5, cmap='YlOrRd', fmt='g', cbar_kws={'label': 'Number of tweets'})
g.set_yticklabels(g.get_yticklabels(), rotation = 0, fontsize = 10)
b, t = plt.ylim()
b += 0.5
t -= 0.5 
plt.ylim(b, t)
plt.title("A Heatmap showing the number of tweets from some areas in Birmingham", fontsize =20)
```
